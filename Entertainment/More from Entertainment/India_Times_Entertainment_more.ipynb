```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'https://www.indiatimes.com/entertainment/bollywood/'
```

```python
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
main = soup.select('.main-container')
```

```python
List = main[0].findAll('li')
```

```python
records = []
for record in List:
    content = record.find('a',{'class':'caption'})
    
    Heading = content.text
    URL = content['href']
    Date = Today_date
    Time = Today_time
    Newspaper='India Times'
    Category = 'bollywood'
    records.append((Newspaper,Heading,Category,Date,Time,URL))
```
