```python
import requests
import bs4
import os
import pandas as pd
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
def record_date_time(z):
    x=str.split(z,',')
    date =x[0]
    time= x[1] +':00'

    return date,time

def record_URL(x):
    y='https://timesofindia.indiatimes.com'
    url = y+x 
    return url
```

```python
link ='https://timesofindia.indiatimes.com/entertainment/hindi/bollywood/news'
```

```python
res = requests.get(link)
```

```python
soup =bs4.BeautifulSoup(res.text,'lxml')
```

```python
toi=soup.find("ul", {"class": "level3"})
```

```python
content = toi.findAll('a',{'data-action': "Navigation_MenuItem"})
```

```python
content
```

```python
def Label(x):
    label = x['data-label']
    if(label=='English'):
        return 'hollywood'
    elif(label=='Hindi'):
        return 'bollywood'
    elif(label== 'Tamil'):
        return 'tamil'
    elif(label =='Telugu'):
        return 'telugu'
    elif(label == 'Malayalam'):
        return 'malayalam'
    else:
        return 'regional'
```

```python
records = []
for labels in content:
    category = Label(labels)
    
    URL = labels['href']
    res = requests.get(URL)
    soup =bs4.BeautifulSoup(res.text,'lxml')
    x = soup.find('div',{'id':'mainlisting'})
    content_labels = x.findAll('div',{'class':'md_news_box'})
    for record in content_labels:
        try:
            Heading = record.text
            URL = record_URL(record.find('a')['href'])
            Date,Time = record_date_time(record.findAll('span')[-1]['dateval'])
            Category = category
            Newspaper = 'Time of India'
        except:
            continue
        records.append((Newspaper,Heading,Category,Date,Time,URL))
```

```python
df=pd.DataFrame(records,columns=['Newspaper','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
