```python
link = 'https://indianexpress.com/section/entertainment/'
```

```python
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
List = soup.findAll('li',{"class":"m-nav-list__item "})
```

```python
List[0].find('a')['href']
```

```python
## bollywood
link = 'https://indianexpress.com/section/entertainment/bollywood/'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')

```

```python
str.split('https://indianexpress.com/section/entertainment/bollywood','/')[-1]
```

```python
h2=soup.select('.m-article-landing__inner')
```

```python
records = []
for page in List:
    link = page.find('a')['href']
    res = requests.get(link)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    category = str.split(link,'/')[-2]
    
    for record in h2:
        content = record.find('h2')
        datetime = record.time.attrs['datetime']

        Heading = content.text
        URL = content.find('a')['href']
        x=str.split(datetime,' ')
        Date = x[0]
        Time = str.split(x[1],'+')[0]
        Newspaper = 'Indian Express'
        Category = category
        records.append((Newspaper,Heading,Category,Date,Time,URL))
```

```python
df=pd.DataFrame(records,columns=['Newspaper','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
