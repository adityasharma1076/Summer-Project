```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link1 = 'https://www.indiatoday.in/movies'
link2 = 'https://www.indiatoday.in/television'
```

```python
res1 = requests.get(link1)
soup1 = bs4.BeautifulSoup(res1.text,'lxml')
```

```python
content = soup1.select(".col-md-4.col-sm-4.col-xs-12.mt-50")
```

```python
def record_URL(x):
    y='https://www.indiatoday.in'
    url = y+x 
    return url
```

```python
content[0].find('a')['href']
```

```python
str.split(content[0].find('a')['href'],'/')[-1]
```

```python
records = []
for List in content:
    List_URL = record_URL(List.find('a')['href'])
    category = str.split(List.find('a')['href'],'/')[-1]
    res = requests.get(List_URL)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    List_content = soup.select('.catagory-listing')
    for record in List_content:
        Heading = record.text
        URL= record_URL(record.find('a')['href'])
        Date = Today_date
        Time = Today_time + ':00'
        Newspaper = 'India Today'
        Category = category
        records.append((Newspaper,Heading,Category,Date,Time,URL))
```

```python
res2 = requests.get(link2)
soup2 = bs4.BeautifulSoup(res2.text,'lxml')
```

```python
content = soup2.select(".itg-listing")
```

```python
tv_news=content[0].findAll('li')
```

```python
for record in tv_news:
    
    Heading = record.text
    URL= record_URL(record.find('a')['href'])
    Date = Today_date
    Time = Today_time
    Newspaper = 'India Today'
    Category = 'television'
    records.append((Newspaper,Heading,Category,Date,Time,URL))
    
```

```python
link3 = 'https://www.indiatoday.in/television/web-series'
```

```python
res3 = requests.get(link3)
soup3 = bs4.BeautifulSoup(res3.text,'lxml')
```

```python
List_content = soup3.select(".catagory-listing")
```

```python
for record in List_content:
        Heading = record.text
        URL= record_URL(record.find('a')['href'])
        Date = Today_date
        Time = Today_time + ':00'
        Newspaper = 'India Today'
        Category = 'web-series'
        records.append((Newspaper,Heading,Category,Date,Time,URL))
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Newspaper','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
