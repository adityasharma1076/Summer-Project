```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
def record_URL(x):
    y='https://www.indiatoday.in'
    url = y+x 
    return url
```

```python
link = 'https://www.indiatoday.in/sports'
```

```python
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
main = soup.select(".col-md-6.col-sm-3.col-xs-3")[0]
```

```python
cat = ['news','hockey','cricket','tennis','athletics','badminton','other sports','football']
```

```python
Content = main.findAll('li')
```

```python
records = []
for List in Content:
    
    category = List.text.lower()[1:]
    if category in cat:
        url = List.find('a')['href']
        if(str.split(url,':')[0]=='https'):
            List_URL = url
        else:
            List_URL = record_URL(url)
        
        res = requests.get(List_URL)
        soup = bs4.BeautifulSoup(res.text,'lxml')
        List_content = soup.select('.catagory-listing')
        for record in List_content:
            Heading = record.text
            URL= record_URL(record.find('a')['href'])
            Date = Today_date
            Time = Today_time + ':00'
            Source = 'India Today'
            Category = category
            records.append((Source,Heading,Category,Date,Time,URL))
    else:
        continue
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.to_csv(fullname,index=False,encoding='utf-8')
```
