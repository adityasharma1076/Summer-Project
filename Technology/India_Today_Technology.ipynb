```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'https://www.indiatoday.in/technology'
```

```python
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
content = soup.select(".itg-listing")
top_news=content[0].findAll('li')
```

```python
featured = soup.select(".featured-news")
featured_news = featured[0].findAll('div')
```

```python
section_news = soup.select(".section-ordering")
```

```python
other = soup.select(".col-md-6.col-sm-6")
```

```python
x=other[3].findAll('li')
```

```python
def record_URL(x):
    y='https://timesofindia.indiatimes.com'
    url = y+x 
    return url
```

```python
records=[]
for record in featured_news:
    
    Heading = record.text
    URL= record_URL(record.find('a')['href'])
    Date = Today_date
    Time = Today_time
    Newspaper = 'India Today'
    records.append((Newspaper,Heading,Date,Time,URL))
```

```python

for record in top_news:
    
    Heading = record.text
    URL= record_URL(record.find('a')['href'])
    Date = Today_date
    Time = Today_time
    Newspaper = 'India Today'
    records.append((Newspaper,Heading,Date,Time,URL))
```

```python
section_news[-2].findAll('p')
```

```python
for i in section_news:
    li = i.findAll('p')
    for record in li:
        Heading = record.text
        URL= record_URL(record.find('a')['href'])
        Date = Today_date
        Time = Today_time
        Newspaper = 'India Today'
        records.append((Newspaper,Heading,Date,Time,URL))
```

```python
for other_news in other:
        li = other_news.findAll('li')

        for record in other_news:
            Heading = record.text
            URL= record_URL(record.find('a')['href'])
            Date = Today_date
            Time = Today_time
            Newspaper = 'India Today'
            records.append((Newspaper,Heading,Date,Time,URL))
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Newspaper','Heading','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
