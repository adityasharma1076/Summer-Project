```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'http://www.theluxecafe.com/'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
List = soup.find('div',{'class':'menu-menu-1-container'}).findAll('ul',{'class':'sub-menu'})
for i in List:
    i.decompose()
```

```python
List = soup.find('div',{'class':'menu-menu-1-container'}).findAll('li',{'class':'menu-item'})
```

```python
List[1].text[:-2]
```

```python
records = []
for link in List[:-1]:
    url = link.find('a')['href']
    category = link.text[:-2].lower()
    res = requests.get(url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    content = soup.findAll('div',{'class':'storyCont'})
    
    for record in content:
        Heading = record.find('img')['title']
        URL = record.find('a')['href']
        Date = Today_date
        Time = Today_time
        Source = 'The Luxe Cafe'
        Category = category
        records.append((Source,Heading,Category,Date,Time,URL))
        
    url = link.find('a')['href']+'page/2/'
    category = link.text[:-2].lower()
    res = requests.get(url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    content = soup.findAll('div',{'class':'storyCont'})
    
    for record in content:
        Heading = record.find('img')['title']
        URL = record.find('a')['href']
        Date = Today_date
        Time = Today_time
        Source = 'The Luxe Cafe'
        Category = category
        records.append((Source,Heading,Category,Date,Time,URL))
        
    
        
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
def fix_cat(x):
    if(x=='life\nvideo'):
        return 'life'
    else:
        return x
df['Category'] = df['Category'].apply(fix_cat)
```

```python
outname ='TLC_lifestyle-'+str(Today_date)+'.csv'

root = 'CSV/'
if not os.path.exists(root):
    os.mkdir(root)
date_today= Today_date +'/'
outdir=root+ date_today[:-1]
if not os.path.exists(outdir):
    os.mkdir(outdir)
fullname = os.path.join(outdir, outname)
```

```python
df.to_csv(fullname, index=False, encoding='utf-8')
```
