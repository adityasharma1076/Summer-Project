```python
import bs4
import requests
import pandas as pd
import datetime
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
def record_URL(x):
    y='http://www.savvy.co.in'
    url = y+x 
    return url
```

```python
link = 'http://www.savvy.co.in/'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
List = soup.find('div',{'class':'subnav'}).findAll('li',{'class':'dropdown'})
for div in List:
    div.decompose()
```

```python
List = soup.find('div',{'class':'subnav'}).findAll('li')
```

```python
List[-1]
```

```python
records = []
for link in List[2:]:
    url = record_URL(link.find('a')['href'])
    category = link.text.lower()
    res = requests.get(url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    content = soup.findAll('div',{'class':'grid-item col25'})
    
    for record in content:
        URL = record_URL(record.find('a')['href'])
        Heading = record.find('p',{'class':'title'}).text
        Date = record.find('span',{'class':'time'}).text
        Time = Today_time
        Category = category
        Source = 'Savvy'
        records.append((Source,Heading,Category,Date,Time,URL))
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
outname ='Savvy_lifestyle-'+str(Today_date)+'.csv'

root = 'CSV/'
if not os.path.exists(root):
    os.mkdir(root)
date_today= Today_date +'/'
outdir=root+ date_today[:-1]
if not os.path.exists(outdir):
    os.mkdir(outdir)
fullname = os.path.join(outdir, outname)
```

```python
df.to_csv(fullname, index=False, encoding='utf-8')
```
