```python
import requests
import bs4
import os
import pandas as pd
```

```python
import datetime
now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'https://www.newsnation.in/india-news/'
```

```python
records = []
for i in range(1,5):
    url = link+str(i)
    res = requests.get(url)
    soup =bs4.BeautifulSoup(res.text,'lxml')
    List = soup.findAll('li',{'class':'w100p'})

    for record in List:
        try:
            URL = record.find('h2').find('a')['href']
            Heading = record.find('h2').text
            Date =  Today_date
            Time = Today_time
            Source = 'Newsnation'
            Category = 'politics'
            records.append((Source,Heading,Category,Date,Time,URL))
        except:
            continue
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
def fix_head(x):
    Heading=x
    while(Heading[0]=='\n'):
        Heading=Heading[1:]
    while(Heading[-1]=='\n'):
        Heading=Heading[:-1]
    while(Heading[0]==' '):
        Heading=Heading[1:]
    return Heading
df['Heading']=df['Heading'].apply(fix_head)
```
