```python
import requests
import bs4
import os
import pandas as pd
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link ='https://www.firstpost.com/category/politics'
res = requests.get(link)
soup =bs4.BeautifulSoup(res.text,'lxml')
```

```python
List = soup.find('div',{'class':"right-content sponsored-side-content"}).findAll('li')
```

```python
records = []
for record in List:
    URL = record.find('a')['href']
    Heading = record.text
    Category = 'politics'
    Date = Today_date
    Time = Today_time
    Source = 'Times of India'
    records.append((Source,Heading,Category,Date,Time,URL))
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
def fix_head(x):
    Heading=x
    while(Heading[0]=='\n'):
        Heading=Heading[1:]
    while(Heading[-1]=='\n'):
        Heading=Heading[:-1]
    while(Heading[0]==' '):
        Heading=Heading[1:]
    return Heading
```

```python
df['Heading']=df['Heading'].apply(fix_head)
```

```python
outname ='FP_politics-'+str(Today_date)+'.csv'

root = 'CSV/'
if not os.path.exists(root):
    os.mkdir(root)
date_today= Today_date +'/'
outdir=root+ date_today[:-1]
if not os.path.exists(outdir):
    os.mkdir(outdir)
fullname = os.path.join(outdir, outname)
```

```python
df.to_csv(fullname,index=False,encoding='utf-8')
```
