```python
import os
import pandas as pd
import numpy as np
import requests
import bs4
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
def record_date_time(z):
    x=str.split(z,' ')
    x[1] = fix_date(x[1])
    date =x[1]+' '+x[2]+' '+x[3]
    time= x[4] +':00'

    return date,time
```

```python
def fix_date(p):
    if(p=='January'):
        return 'Jan'
    elif(p=='February'):
        return 'Feb'
    elif(p=='March'):
        return 'Mar'
    elif(p=='April'):
        return 'Apr'
    elif(p=='May'):
        return 'May'
    elif(p=='June'):
        return 'Jun'
    elif(p=='July'):
        return 'Jul'
    elif(p=='August'):
        return 'Aug'
    elif(p=='September'):
        return 'Sep'
    elif(p=='October'):
        return 'Oct'
    elif(p=='November'):
        return 'Nov'
    elif(p=='Dececmber'):
        return 'Dec'
```

```python
link = 'https://www.thehindu.com/education/'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
scrape = soup.find('nav',{"class":"navbar sub-menu-sticky hidden-xs "})
```

```python
List = scrape.findAll('li')
```

```python
def fix_cat(x):    
    if(x=='Other Sports '):
        return 'others'
    else:
        return x.lower()
```

```python
records = []
for link in List[1:]:
    url=link.find('a')['href']
    category = fix_cat(link.text)
    res = requests.get(url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
  
    maincontent = soup.find('div',{'class':'story1-3x100-container s1-3x100BlueBg-container hover-icon'}).findAll('h2')
    for record in maincontent:


        URL = record.find('a')['href']
        Heading = record.text[2:-2]

        Date,Time = record_date_time(record.find('a')['title'])
        Category = category
        Source = 'The Hindu'
        records.append((Source,Heading,Category,Date,Time,URL))
    
    

    maincontent = soup.findAll('div',{'class':'story-card'})
    for record in maincontent[0:]:


        z=record.find('h2')
        if not z:
            z=record.find('h3')

        URL = z.find('a')['href']
        Heading = z.text[2:-2]

        Date,Time = record_date_time(z.find('a')['title'])
        Category = category
        Source = 'The Hindu'
        records.append((Source,Heading,Category,Date,Time,URL))

    
    maincontent = soup.findAll('div',{'class':'Other-StoryCard'})
    for record in maincontent[1:10]:
        z=record.find('h3')
            
        URL = z.find('a')['href']
        Heading = z.text[2:-2]

        Date,Time = record_date_time(z.find('a')['title'])
        Category = category
        Source = 'The Hindu'
        records.append((Source,Heading,Category,Date,Time,URL))
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
