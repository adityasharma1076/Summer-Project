```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'https://www.hindustantimes.com/education/'
```

```python
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
List = soup.find('div',{'class' : "col-sm-7 col-md-8 col-lg-9"}).findAll('div',{'class':'media-body'})
```

```python
List[0]
```

```python
records = []
for record in List:
    Heading = record.find('a').text
    URL = record.find('a')['href']
    Source = 'Hindiustan Times'
    dt=str.split(record.find('span').text)
    Date = dt[0]+' '+dt[1]+' '+dt[2]
    Time = dt[3]+':00'
    Category = 'nil'
    records.append((Source,Heading,Category,Date,Time,URL))
    
```

```python
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
outname ='HT_education-'+str(Today_date)+'.csv'
root = 'CSV/'
if not os.path.exists(root):
    os.mkdir(root)
date_today= Today_date +'/'
outdir=root+ date_today[:-1]
if not os.path.exists(outdir):
    os.mkdir(outdir)
fullname = os.path.join(outdir, outname)
```

```python
df.to_csv(fullname,index=False,encoding='utf-8')
```
