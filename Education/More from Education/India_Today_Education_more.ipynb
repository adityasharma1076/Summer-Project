```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'https://www.indiatoday.in/education-today'
```

```python
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
content = soup.select(".col-md-4.col-sm-6.col-xs-12.mt-50")
```

```python
def record_URL(x):
    y='https://www.indiatoday.in'
    url = y+x 
    return url
```

```python
records = []
for List in content:
    List_URL = record_URL(List.find('a')['href'])
    category = str.split(List.find('a')['href'],'/')[-1]
    res = requests.get(List_URL)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    List_content = soup.select('.catagory-listing')
    for record in List_content:
        Heading = record.text
        URL= record_URL(record.find('a')['href'])
        Date = Today_date
        Time = Today_time + ':00'
        Source = 'India Today'
        Category = category
        records.append((Source,Heading,Category,Date,Time,URL))
        
    List_URL = record_URL(List.find('a')['href'])+'?page=1'
    category = str.split(List.find('a')['href'],'/')[-1]
    res = requests.get(List_URL)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    List_content = soup.select('.catagory-listing')
    for record in List_content:
        Heading = record.text
        URL= record_URL(record.find('a')['href'])
        Date = Today_date
        Time = Today_time + ':00'
        Source = 'India Today'
        Category = category
        records.append((Source,Heading,Category,Date,Time,URL))   
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
outname ='IT_education-'+str(Today_date)+'.csv'
root = 'CSV/'
if not os.path.exists(root):
    os.mkdir(root)
date_today= Today_date +'/'
outdir=root+ date_today[:-1]
if not os.path.exists(outdir):
    os.mkdir(outdir)
fullname = os.path.join(outdir, outname)
```

```python
df.to_csv(fullname,index=False,encoding='utf-8')
```
