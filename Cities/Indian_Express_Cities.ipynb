```python
import bs4
import requests
import pandas as pd
import numpy as np
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
def get_date_time(URL):
    req = requests.get(URL)
    soup_date = bs4.BeautifulSoup(req.text,'lxml')
    temp = soup_date.select(".m-story-meta__credit")
    time_date = temp[0].findAll('span')[-1].text
    x=str.split(time_date)
    date = x[1]+' '+x[2]+' '+x[3]
    time = x[4]+' '+x[5]
    return date,time
```

```python
link = 'https://indianexpress.com/section/cities/'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
h1 = soup.select('.m-article-featured__title')
```

```python
records=[]
h1_newspaper='Indian Express'
h1_heading = h1[0].text
h1_url = h1[0].find('a')['href']
h1_date,h1_time = get_date_time(h1_url)
h1_category = 'cities'

records.append((h1_newspaper,h1_heading,h1_category,h1_date,h1_time,h1_url))
```

```python
h2=soup.findAll('h2')
```

```python
for record in h2[1:-2]:
    Heading = record.text
    URL = record.find('a')['href']
    Date,Time = get_date_time(URL)
    #Time = get_time(URL)
    Source='Indian Express'
    Category = 'cities'
    records.append((Source,Heading,Category,Date,Time,URL))
```

```python
h3=soup.select('.m-article-landing__inner')
```

```python
for record in h3:
    content = record.select(".m-article-landing__title")[0]
    datetime = record.time.attrs['datetime']

    Heading = content.text
    URL = content.find('a')['href']
    x=str.split(datetime,' ')
    Date = x[0]
    Time = str.split(x[1],'+')[0]+':00'
    Source = 'Indian Express'
    Category = 'cities'

    records.append((Source,Heading,Category,Date,Time,URL))
```

```python
List = soup.findAll('li',{'class':'m-nav-list__item '})
```

```python
for link in List:
    category= link.text.lower()+' news'
    url = link.find('a')['href']
    res = requests.get(url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    maincontent = soup.findAll('div',{'class' : 'm-article-landing__inner'})
    
    for record in maincontent:
        content = record.select(".m-article-landing__title")[0]
        datetime = record.time.attrs['datetime']

        Heading = content.text
        URL = content.find('a')['href']
        x=str.split(datetime,' ')
        Date = x[0]
        Time = str.split(x[1],'+')[0]+':00'
        Source = 'Indian Express'
        Category = category
        records.append((Source,Heading,Category,Date,Time,URL))
```

```python
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
