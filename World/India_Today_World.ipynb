```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'https://www.indiatoday.in/world'
```

```python
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
content = soup.select(".view-content")
```

```python
top_news=content[0].findAll("div",{"class":'catagory-listing'})
```

```python
def record_URL(x):
    y='https://www.indiatoday.in'
    url = y+x 
    return url
```

```python
records=[]
for record in top_news:
    
    Heading = record.text
    URL= record_URL(record.find('a')['href'])
    Date = Today_date
    Time = Today_time
    Newspaper = 'India Today'
    Category = 'World'
    records.append((Newspaper,Heading,Category,Date,Time,URL))
```

```python
# Page 2
link = 'https://www.indiatoday.in/world?page=1'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
content = soup.select(".view-content")
top_news=content[0].findAll("div",{"class":'catagory-listing'})
for record in top_news:
    
    Heading = record.text
    URL= record_URL(record.find('a')['href'])
    Date = Today_date
    Time = Today_time
    Newspaper = 'India Today'
    Category = 'World'
    records.append((Newspaper,Heading,Category,Date,Time,URL))
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Newspaper','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
