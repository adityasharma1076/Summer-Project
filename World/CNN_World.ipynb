```python
import bs4
import requests
import os
import pandas as pd
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link ='https://edition.cnn.com/world'
res = requests.get(link)
soup =bs4.BeautifulSoup(res.text,'lxml')
```

```python
main = soup.find('div',{'id':'nav-section-submenu'})
List = main.findAll('a')
```

```python
def record_URL(x):
    y='https://edition.cnn.com'
    url = y+x 
    return url
```

```python
List[0].text
```

```python
records = []
for i in List:
    url = record_URL(i['href'])
    category = i.text.lower()
    res = requests.get(url)
    soup =bs4.BeautifulSoup(res.text,'lxml')
    content = soup.findAll('h3',{'class' : 'cd__headline'})
    
    for record in content:
        Heading = record.text
        URL = record_URL(record.find('a')['href'])
        Date = Today_date
        Time = Today_time
        Category = category
        Source = 'CNN'
        
        records.append((Source,Heading,Category,Date,Time,URL))
        
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
outname ='CNN_world-'+str(Today_date)+'.csv'
x=r"C:\Users\drago\Documents\GitHub\Summer-Project\World\CSV\ "
if not os.path.exists(x[:-1]):
    os.mkdir(x[:-1])
date_today= Today_date +'\ '
outdir=x[:-1]+ date_today[:-1]
if not os.path.exists(outdir):
    os.mkdir(outdir)
fullname = os.path.join(outdir, outname)
```

```python
df.to_csv(fullname,index=False,encoding='utf-8')
```
