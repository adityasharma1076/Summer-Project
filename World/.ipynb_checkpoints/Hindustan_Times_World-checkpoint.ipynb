```python
import requests
import bs4
import pandas as pd
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'https://www.hindustantimes.com/world-news/'
```

```python
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
List = soup.select('.row.pb-20')
```

```python
links = List[0].findAll('h2')
links[0].text[1:-1]
```

```python
for i in links:
    print(i.text[1:-1])
```

```python
def fix_category(x):
    x=x[1:-1]
    if (x=='Cricket News'):
        return 'cricket'
    elif(x =='Football News'):
        return 'football'
    elif(x =='Tennis News'):
        return 'tennis'
    else:
        return 'other'
```

```python
records = []
for link in links:
    url = link.find('a')['href']
    category = fix_category(link.text)
    res = requests.get(url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    content = soup.findAll('div',{'class':'media-body'})
    for record in content:
        Heading = record.find('a').text
        URL = record.find('a')['href']
        Newspaper = 'Hindiustan Times'
        dt=str.split(record.find('span').text)
        Date = dt[0]+' '+dt[1]+' '+dt[2]
        Time = dt[3]+':00'
        Category = category
        records.append((Newspaper,Heading,Category,Date,Time,URL))
    
```

```python
df=pd.DataFrame(records,columns=['Newspaper','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
