```python
import requests
import bs4
import os
import pandas as pd
import numpy as np
import datetime
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
def record_date_time(z):
    x=str.split(z,',')
    date =x[0]
    time= x[1] +':00'

    return date,time

def record_URL(x):
    y='https://www.businesstoday.in'
    url = y+x 
    return url
```

```python
link = 'https://www.businesstoday.in/'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
scrape=soup.find('div',{'class' : 'menues'}).findAll('li')
```

```python
records = []
for link in scrape[2:]:
    for i in range(1,5):
        url = record_URL(link.find('a')['href'])+'?page='+str(i)
        category = link.find('a').text.lower()
        res = requests.get(url)
        soup = bs4.BeautifulSoup(res.text,'lxml')
        content = soup.findAll('div',{'class':'boxcont'})
        for record in content:
            try:
                Heading = record.attrs['ctitle']
                URL = record_URL(record.attrs['url'])
                Date = Today_date
                Time = Today_time
                Category = category
                Source = 'Business Today'
                records.append((Source,Heading,Category,Date,Time,URL))
            except:
                continue
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.to_csv(fullname,index=False,encoding='utf-8')
```
