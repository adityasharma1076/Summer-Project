```python
import os
import pandas as pd
import numpy as np
import requests
import bs4
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
def record_date_time(z):
    x=str.split(z,' ')
    x[1] = fix_date(x[1])
    date =x[1]+' '+x[2]+' '+x[3]
    time= x[4] +':00'

    return date,time
```

```python
def fix_date(p):
    if(p=='January'):
        return 'Jan'
    elif(p=='February'):
        return 'Feb'
    elif(p=='March'):
        return 'Mar'
    elif(p=='April'):
        return 'Apr'
    elif(p=='May'):
        return 'May'
    elif(p=='June'):
        return 'Jun'
    elif(p=='July'):
        return 'Jul'
    elif(p=='August'):
        return 'Aug'
    elif(p=='September'):
        return 'Sep'
    elif(p=='October'):
        return 'Oct'
    elif(p=='November'):
        return 'Nov'
    elif(p=='December'):
        return 'Dec'
```

```python
link = 'https://www.thehindu.com/business/'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
links = soup.find('nav',{'class':"navbar sub-menu-sticky hidden-xs "}).findAll('li')
```

```python
records = []
for link in links:
    url = link.find('a')['href']
    category = link.text.lower()
    
    res = requests.get(url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    List = soup.findAll('div',{'class':'story-card-33'})
    for record in List:
        z=record.find('p')

        URL = z.find('a')['href']
        Heading = z.text[4:-4]

        Date,Time = record_date_time(z.find('a')['title'])
        Category = category
        Source = 'The Hindu'
        records.append((Source,Heading,Category,Date,Time,URL))


    maincontent = soup.findAll('div',{'class':'Other-StoryCard'})    
    for record in maincontent[1:30]:


        z=record.find('h3')

        URL = z.find('a')['href']
        Heading = z.text[2:-2]

        Date,Time = record_date_time(z.find('a')['title'])
        Category = category
        Source = 'The Hindu'
        records.append((Source,Heading,Category,Date,Time,URL))
```

```python
import pandas as pd
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
```
