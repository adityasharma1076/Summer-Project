```python
import requests
import bs4
import pandas as pd
import datetime
import numpy as np
import os
```

```python
import datetime

now = datetime.datetime.now()


Today_time = now.strftime("%H:%M")

Today_date = now.strftime("%Y-%m-%d")
```

```python
link = 'https://www.cnbctv18.com/'
res = requests.get(link)
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
scrape = soup.find('ul',{'class':'menu overlay-content'}).find('div',{'id':'news'})
```

```python
List = scrape.findAll('a')
```

```python
List
```

```python
records = []
for link in List:
    category = link.text.lower()
    for i in range(1, 3):
        url = link.attrs['href']+'page-'+str(i)
        res = requests.get(url)
        soup = bs4.BeautifulSoup(res.text,'lxml')
        content = soup.findAll('div',{'class':'list_title'})
        for record in content:
            URL = record.find('a')['href']
            Heading = record.text
            while(Heading[-1]=='\n'):
                Heading=Heading[:-1]
            Category = category
            Date =Today_date
            Time = Today_time
            Source = 'cnbc'
            
            records.append((Source,Heading,Category,Date,Time,URL))
    
    
```

```python
df=pd.DataFrame(records,columns=['Source','Heading','Category','Date','Time','URL'])
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
outname ='CNBC_business-'+str(Today_date)+'.csv'

root = 'CSV/'
if not os.path.exists(root):
    os.mkdir(root)
date_today= Today_date +'/'
outdir=root+ date_today[:-1]
if not os.path.exists(outdir):
    os.mkdir(outdir)
fullname = os.path.join(outdir, outname)
```

```python
df.to_csv(fullname,index=False,encoding='utf-8')
```
